```python
import pandas as pd
def categorize_age(age):
    if pd.isnull(age):
        return 'unknown'
    elif 13<= age <= 20:
        return 'teenagers'
    elif 21 <= age <= 40:
        return 'elder agers'
    elif 41<= age <= 65:
         return 'Senior Citizens'
    else:
        return 'Child'

df['Age_group'] = df['Age'].apply(categorize_age)
    
```

```python
age_group_p = df['Age_group'].value_counts()

plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(age_group_p, labels=age_group_p.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Age Groups")
plt.show()
```

```python
age_group_counts = df['Age Group'].value_counts()

plt.figure(figsize=(8, 6))  # Set figure size
age_group_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Distribution of Age Groups', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Count', fontsize=14)

```

```python
# Pivot Table for Product Sales by Region
pivot_table = df.pivot_table(values='Total Price', index='Region', columns='Product Name', aggfunc='sum', fill_value=0)
print(pivot_table)
```

```python
import matplotlib.pyplot as plt

# Example: Product-wise Total Sales
df.groupby('Product Name')['Total Price'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by Region and sum Total Price
df.groupby('Region')['Total Price'].sum().plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

```
